```python
def get_all_users( json_str = False ):
    conn.row_factory = sqlite3.Row # This enables column access by name: row['column_name'] 
    db = conn.cursor()

    rows = db.execute('''
    SELECT * from tweet
    ''').fetchall()

    conn.commit()
    conn.close()

    if json_str:
        return [dict(ix) for ix in rows] #CREATE JSON

    return rows

twtdata = get_all_users(json_str = True)
```
