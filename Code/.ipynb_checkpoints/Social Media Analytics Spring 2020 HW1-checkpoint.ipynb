```python
#Homework 1 - Graph Theory Basics
import json,random,csv
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Question 1:  Load the retweet graph that will be the basis for this question
with open('starwars/bsw-0(1).json') as json_file:
    data = json.load(json_file)
    c=0
    e=0
    key = 'retweeted_status'
    EdgeList = []
    print('Dataset of %s tweets'%len(data))
    cmax = min(1000,len(data))-1
    G = nx.DiGraph()
    while c<=cmax:        
        tweet = data[c]
        c+=1
        if key in tweet.keys():
            e +=1
            retweeter = tweet['user']['screen_name']
            source = tweet[key]['user']['screen_name']
            text = tweet[key]['text']
            #print('Edge %s (%s,%s): %s'%(e,source,retweeter,text_dict['text']))
            EdgeList.append((source,retweeter))
            G.add_node(source,text=text)
            G.add_node(retweeter,text="RT @"+source+": "+text)
            G.add_edge(source,retweeter)
nv = G.number_of_nodes()
ne = G.number_of_edges()
print('Retweet network has %s nodes and %s edges'%(nv,ne))
```

```python
#Question 1.1 
#Draw the retweet network.  Choose the node, edge, and background color to make it look cool.
fig = plt.figure()
Gu = G.to_undirected(G)
pos=nx.random_layout(Gu)
nx.draw(Gu,pos, width=0.25,node_color="purple",edge_color="purple",node_size=5)
fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 1.2
#How many edges and nodes are in the graph?
print("Graph has %s edges"%G.number_of_edges())
print("Graph has %s nodes"%G.number_of_nodes())
```

```python
#Question 1.3
#Find the nodes with the top three highest out-degrees.  What are their name, out-degree, and corresponding tweet? 
dsort = sorted(list(G.out_degree()), key = lambda x: x[1],reverse = True)
for x in dsort[0:3]:
    user = x[0]
    dout = x[1]
    tweet = G.nodes[user]['text']
    print('%s retweets: %s: %s'%(dout,user,tweet))

```

```python
#Question 1.4
#Find the nodes with the top three highest in-degrees.  What are their name, in-degree, and corresponding tweet? 
dsort = sorted(list(G.in_degree()), key = lambda x: x[1],reverse = True)
for x in dsort[0:3]:
    user = x[0]
    din = x[1]
    tweet = G.nodes[user]['text']
    print('%s retweets: %s: %s'%(din,user,tweet))
```

```python
#Question 1.5
#Is in-degree or out-degree a better measure of importance in the retweet network?  Why?
print("Out-degree is a better measure because it captures who has the most retweets.")
```

```python
#Question 1.6
#Redraw the retweet network, but now make the node size proportional to the out-degree.
dout = dict(G.out_degree())
fig = plt.figure()
pos = nx.spring_layout(Gu,k=0.1)
nx.draw(Gu,pos, nodelist=list(dout.keys()), node_size=[v+1  for v in dout.values()],width=0.25,node_color="purple",edge_color="purple")
fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 2
#Load the rumor network for this question.  
Nodes ={}
Edges =[]
with open('rumorgraph.nodes.csv') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    c=0
    for row in data:
        if c>=1:
            Nodes[row[0]]= row[1]
        c+=1

with open('rumorgraph.edges.csv') as csvfile:
    data = csv.reader(csvfile, delimiter=',')
    c=0
    for row in data:
        if c>=1:
            Edges.append((Nodes[row[0]],Nodes[row[1]]))
        c+=1
G = nx.DiGraph()
for edge in Edges:
    source = edge[0]
    recipient = edge[1]    
    G.add_node(source)
    G.add_node(recipient)
    G.add_edge(source,recipient)
G.add_edge("Tauhid's Mom","Tulip's Mom")

nv = G.number_of_nodes()
ne = G.number_of_edges()
print('Rumor network has %s nodes and %s edges'%(nv,ne))
```

```python
#Question 2.1
#Draw the rumor graph with nodes labeled by their names.
#Choose the node, edge, and background color to make it look cool.

fig = plt.figure()
pos=nx.spring_layout(G,k=.4)
r=0.01
for u in pos.keys():
    xy = pos[u]
    pos[u] = (xy[0]+r*random.gauss(0,1),xy[1]+r*random.gauss(0,1))
nx.draw(G, width=2,pos=pos,node_color="red",edge_color="green",node_size=150)
nx.draw_networkx_labels(G, pos,font_size=8,font_color="white")
fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 2.2
#Find any cycles in the graph and list their edges. 
nx.find_cycle(G, orientation='original')
```
