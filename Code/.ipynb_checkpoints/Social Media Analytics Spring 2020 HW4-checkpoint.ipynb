```python
#Homework 4 - Visualizing big graphs
import json,random,csv
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path = "C:/Users/Zlisto/Dropbox (MIT)/Opinion_Dynamics/Social_Network_Real_Data/USPresidentialDebate/"
nodes = pd.read_csv(path+"polarities.csv") 
print("network has %s nodes"%len(nodes))
nodes.head()

```

```python

dmin = 1e5
nodes_keep = nodes.loc[nodes['n_followers']>=dmin]['id'].to_list();
nodes_keep  = set(nodes_keep)
print("%s nodes have more than %s followers"%(len(nodes_keep),dmin))
```

```python
G=nx.DiGraph()
vmax = len(nodes_keep)
with open(path+"friend_graph.csv") as f:
    line = f.readline()
    nv = 0
    ne = 0
    while line:
        line = f.readline()
        #print(line[0:20])
        try:
            line_list = [float(x) for x in line.split(",")]
            v = line_list[0]
            nb = line_list[1:]
            if v in nodes_keep:
                nv+=1
                if nv%10==0:
                    print("Node %s: %s dout = %s"%(nv,nodes.loc[nodes['id']==v,'screen_name'].to_list()[0],len(nb))) 
                for u in nb:
                    if u in nodes_keep:
                        ne+=1                       
                        G.add_edge(v,u)
                        #print("\t Target: %s"%(nodes.loc[nodes['id']==u,'screen_name'].to_list()[0])) 

                                #print("\tEdge %s: (%s,%s)"%(ne,v_screen_name['screen_name'].to_list()[0]
                                #                            ,u_screen_name['screen_name'].to_list()[0]))

            if nv>vmax:
                break
        except:
            print("Error with: %s"%line)
nv = G.number_of_nodes()
ne = G.number_of_edges()
print("Done loading graph with %s nodes, %s edges"%(nv,ne))
nx.write_gpickle(G, "twitter_network_dmin_%s"%(dmin))
nx.write_edgelist(G, "twitter_edgelist_dmin_%s.txt"%(dmin))
```

```python
#Question 1
#We are going to draw a big network.  Probably the biggest you have seen in your life.
#Load up the network 
fname = "twitter_network_dmin_100000.0"
G = nx.read_gpickle(fname)
nv = G.number_of_nodes()
ne = G.number_of_edges()
print('Retweet network has %s nodes and %s edges'%(nv,ne))
print('Network has %s nodes and %s edges'%(nv,ne))
```

```python
#Question 1.1
#Plot the network using a random
Gu = G.to_undirected()
fig = plt.figure()
pos=nx.random_layout(Gu)

node_color = []
node_size =[]

# for each node in the graph
for node in G.nodes(data=False):
    node_size.append(max(10,G.out_degree(node)))
    node_color.append('purple')
nx.draw(Gu, width=0.1,pos=pos,node_color=node_color,edge_color="green",node_size=node_size,title="Twitter network",connectionstyle='arc3', rad=.03)
#nx.draw_networkx_labels(G, pos,font_size=8,font_color="white")

fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 1.2
#Draw the network with the same layout, but now
#add in the twitter screen names for the nodes with the top 5 out-degree
dout = [(v,G.out_degree(v)) for v in G.nodes()]
dout_sorted = sorted(dout, key=lambda x: x[1],reverse=True)
dout_max_label =  dout_sorted[10][1]
Labels={}
for v in G.nodes():
    if G.out_degree(v)>dout_max_label:
        Labels[v]= G.nodes[v]
    else:
        Labels[v]=''

fig = plt.figure()
node_color = []
node_size =[]
size_nonstubborn = 20
size_stubborn = 100
# for each node in the graph
for node in G.nodes(data=False):
    node_size.append(G.out_degree(node))
    node_color.append('purple')

nx.draw(Gu, width=0.5,pos=pos,node_color=node_color,edge_color="green",node_size=node_size,title="Twitter network")
nx.draw_networkx_labels(G, pos,Labels,font_size=8,font_color="white")

fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 1.3 
#Now lets try a nicer layout.  Draw the network as you did in Question 1.1, but this time 
#use the spring based layout.  Approximately how long did it take to draw the network?
Gu = G.to_undirected()
fig = plt.figure()
pos=nx.spring_layout(Gu,k=.4)

node_color = []
node_size =[]

# for each node in the graph
for node in G.nodes(data=False):
    node_size.append(G.out_degree(node)*.1)
    node_color.append('purple')

nx.draw(Gu, width=0.5,pos=pos,node_color=node_color,edge_color="green",
        node_size=node_size,
        title="Twitter network",
        connectionstyle='arc3', rad=.03)
nx.draw_networkx_labels(G, pos,Labels,font_size=8,font_color="white")

fig.set_facecolor("#00000F")
plt.show()
```

```python
#Question 1.4 
#Now lets try a spectral layout.  Draw the network as you did in Question 1.1, but this time 
#use the spectral based layout.  Approximately how long did it take to draw the network?
Gu = G.to_undirected()
fig = plt.figure()
pos=nx.spectral_layout(Gu,scale=10)

node_color = []
node_size =[]

# for each node in the graph
for node in G.nodes(data=False):
    node_size.append(G.out_degree(node)*.1)
    node_color.append('purple')

nx.draw(Gu, width=.5,pos=pos,node_color=node_color,edge_color="green",node_size=node_size,title="Twitter network",
        connectionstyle='arc3', rad=.03)
nx.draw_networkx_labels(G, pos,Labels,font_size=8,font_color="white")
fig.set_facecolor("#00000F")
plt.show()
```

```python
C = nx.connected_components(Gu)

largest_cc = max(C, key=len)
fig = plt.figure()
Gsub = G.subgraph(largest_cc)
Gu = Gsub.to_undirected()
#pos=nx.spectral_layout(Gu,scale=1)
#pos=nx.spring_layout(Gu,k=.4)
#pos = nx.kamada_kawai_layout(Gu)
#pos = nx.circular_layout(Gu)
#pos = nx.spiral_layout(Gu)
pos = nx.shell_layout(Gu)

node_color = []
node_size =[]

# for each node in the graph
for node in Gu.nodes(data=False):
    node_size.append(Gsub.out_degree(node)*.1)
    node_color.append('purple')

nx.draw(Gu, width=0.1,pos=pos,node_color=node_color,edge_color="green",node_size=node_size,title="Twitter network",
        connectionstyle='arc3', rad=.03)
#nx.draw_networkx_labels(Gsub, pos,Labels,font_size=8,font_color="white")

fig.set_facecolor("#00000F")
plt.show()
```
